### 使用scikit-learn 做SVM分类实例

使用sklearn 本身自带的鸢尾花数据集做分类测试，测试使用不同的核函数下分类的效果

注：该数据集总共有一百五十个鸢尾花的一些尺寸的观测值：萼片长度、宽度，花瓣长度和宽度。还有它们的亚属：山鸢尾（Iris setosa）、变色鸢尾（Iris versicolor）和维吉尼亚鸢尾（Iris virginica）

导入需要的package

```python
from sklearn import datasets
import numpy as np 
from sklearn import model_selection
iris=datasets.load_iris()
```

查看下数据,总共150行4列

```python
iris.data.shape
```

```python
iris.data[0:5]
```

```python
iris.target[0:5]
```

把数据分隔成训练集和测试集

```python
X_train,X_test,y_train,y_test=model_selection.train_test_split(iris.data,iris.target,test_size=0.3,random_state=10)
```

### 学习和预测

标准的sklearn的训练和预测过程  这里使用线性SVM作为例子

```python
from sklearn import svm
```

```python
clf=svm.LinearSVC()
```

```python
clf.fit(X_train,y_train)
```

```python
y_pre=clf.predict(X_test)
y_pre
```

我们可以比较下 预测的 和真实的差距

```python
y_test
```

查看预测分数,可以看到预测效果还算不错

```python
clf.score(X_test,y_test)
```

输出预测的混淆矩阵,可以很直观的看到预测情况,列为真实的标签,行为预测的标签,当只有对角线有数字时,全部预测中,下面的情况表示2个预测错误,把两个真实为1的预测成2了

```python
from sklearn import metrics
metrics.confusion_matrix(y_test,y_pre)
```

使用knn 比较分类效果

```python
from sklearn import neighbors
knn=neighbors.KNeighborsClassifier()
knn.fit(X_train,y_train)
print(knn.score(X_test,y_test))
metrics.confusion_matrix(y_test,knn.predict(X_test))
```

可以看到在当前例子里 knn的分类效果比 线性svm好,但是并不能说SVM就一定不如KNN

### 比较不同内核SVM的分类效果

使用使用sklearn的 svm.SVC分类器,其表示支持向量分类器(support vector classifier) 与之相对应的是 SVR  svm的回归模型

类别不总是可以用线性可分，所以人们指望有些可能是多项式或指数实例的非线性决策函数：

线性核

svc = svm.SVC(kernel=’linear’)

多项式核

svc = svm.SVC(kernel=’poly’, … degree=3) # degree: polynomial degree

RBF核(径向基函数)(SVC默认的核函数)

svc = svm.SVC(kernel=’rbf’) # gamma: inverse of size of # radial kernel

sigmoid核

svc = svm.SVC(kernel=’sigmoid’)

也可以自己定制内核:

具体看:http://scikit-learn.org/stable/modules/svm.html#svm-kernels

例子:http://scikit-learn.org/stable/auto_examples/svm/plot_custom_kernel.html#sphx-glr-auto-examples-svm-plot-custom-kernel-py

```python
print(__doc__)

import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm, datasets

# import some data to play with
iris = datasets.load_iris()
X = iris.data[:, :2]  # we only take the first two features. We could
                      # avoid this ugly slicing by using a two-dim dataset
Y = iris.target


def my_kernel(X, Y):
    """
    We create a custom kernel:

                 (2  0)
    k(X, Y) = X  (    ) Y.T
                 (0  1)
    """
    M = np.array([[2, 0], [0, 1.0]])
    return np.dot(np.dot(X, M), Y.T)
def SVM_kernel(kernel):

    h = .02  # step size in the mesh

    # we create an instance of SVM and fit out data.
    clf = svm.SVC(kernel=kernel)
    clf.fit(X, Y)

    # Plot the decision boundary. For that, we will assign a color to each
    # point in the mesh [x_min, x_max]x[y_min, y_max].
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])

    # Put the result into a color plot
    Z = Z.reshape(xx.shape)
    plt.pcolormesh(xx, yy, Z, cmap=plt.cm.Paired)

    # Plot also the training points
    plt.scatter(X[:, 0], X[:, 1], c=Y, cmap=plt.cm.Paired, edgecolors='k')
    plt.title('3-Class classification using Support Vector Machine with '
              ' kernel')
    plt.axis('tight')
    plt.show()
```

线性核

```python
SVM_kernel("linear")
```

多项式核

```python
SVM_kernel("poly")
```

RBF核(径向基函数)(SVC默认的核函数)

```python
SVM_kernel("rbf")
```

自己定制的核

```python
SVM_kernel(my_kernel)
```
